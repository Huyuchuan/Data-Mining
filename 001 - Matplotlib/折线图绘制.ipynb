# 折线图的绘制与显示
## 举例:展现上海一周的天气,比如从星期一到星期日的天气温度如下

```python
import matplotlib.pyplot as plt

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 2. 绘制图像
plt.plot([1, 2, 3, 4, 5, 6 ,7], [17,17,18,15,11,11,13])

# 3. 保存图像，注意，不能保出现在 `plt.show()` 后面
plt.savefig('pic/折线图.png')

# 4. 显示图像，同时释放画布的所有资源
plt.show()
```

## 完善原始折线图 — 给图形添加辅助功能

```python
# 需求:画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
from matplotlib import pyplot as plt
import random

# 1. 获取数据
x = range(60)  # 创建一个整数列表
y_shanghai = [random.uniform(15, 18) for i in x]

# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3. 绘制图像
plt.plot(x, y_shanghai)

# 修改 x、y 刻度
# 准备 x 的刻度说明
x_lable = ['11点{}分'.format(i) for i in x]
plt.xticks(x[::5], x_lable[::5])  # 每隔 5 是一个单位
plt.yticks(range(0, 40, 5))  # 0 到 40，每隔 5 是一个单位

# 添加描述信息
plt.title('某城市11-12点温度变化')
plt.xlabel('时间')
plt.ylabel('温度')

# 添加网格显示
plt.grid(True, linestyle='--', alpha=0.5)

# 4. 显示图像
plt.show()
```

## 在一个坐标系中绘制多个图像

```python
# 需求:画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
from matplotlib import pyplot as plt
import random

# 1. ------------------ 获取数据 ------------------ 
x = range(60)  # 创建一个整数列表
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(0, 3) for i in x]


# 2. ------------------ 创建画布 ------------------ 
plt.figure(figsize=(20, 8), dpi=80)


# 3. ------------------ 绘制图像 ------------------ 
plt.plot(x, y_shanghai, color='b', linestyle='--', label='上海')
plt.plot(x, y_beijing, color='r', linestyle='-.', label='北京')

# 【显示图例】使上面的 label='上海' 显示出来：
# 因为上面是在图像层绘制的，而 label 需要在**辅助显示层**显示
plt.legend()

# 准备 x 的刻度说明
x_label = ['11点{}分'.format(i) for i in x]

# 修改 x、y 刻度
plt.xticks(x[::5], x_label[::5])  # 每隔 5 是一个单位
plt.yticks(range(0, 40, 5))  # 0 到 40，每隔 5 是一个单位


#  ------------------ 添加描述信息 ------------------ 
plt.title('某城市11-12点温度变化')
plt.xlabel('时间')
plt.ylabel('温度')


#  ------------------ 添加网格显示 ------------------ 
plt.grid(True, linestyle='--', alpha=0.5)


# 4.  ------------------ 显示图像 ------------------ 
plt.show()
```

## **多个坐标系显示 plt.subplots(面向对象的画图方法)**

```python
# 需求:画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
from matplotlib import pyplot as plt
import random

# 1. ------------------ 获取数据 ------------------ 
x = range(60)  # 创建一个整数列表
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(0, 3) for i in x]


# 2. ------------------ 创建画布 ------------------ 
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=80)


# 3. ------------------ 绘制图像 ------------------ 
axes[0].plot(x, y_shanghai, color='b', linestyle='--', label='上海')
axes[1].plot(x, y_beijing, color='r', linestyle='-.', label='北京')

# 【显示图例】使上面的 label='上海' 显示出来：
# 因为上面是在图像层绘制的，而 label 需要在**辅助显示层**显示
axes[0].legend()
axes[1].legend()

# 准备 x 的刻度说明
x_label = ['11点{}分'.format(i) for i in x]

# 修改 x、y 刻度
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::5])
axes[0].set_xticks(x[::5],)  # 每隔 5 是一个单位
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0, 40, 5))  # 0 到 40，每隔 5 是一个单位

axes[1].set_xticks(x[::5], x_lable[::5])  # 每隔 5 是一个单位
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0, 40, 5))  # 0 到 40，每隔 5 是一个单位

#  ------------------ 添加描述信息 ------------------ 
axes[0].set_title('上海11-12点温度变化')
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[1].set_title('北京11-12点温度变化')
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')


#  ------------------ 添加网格显示 ------------------ 
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)


# 4.  ------------------ 显示图像 ------------------ 
plt.show()
```
